```python
# extracting the urls
urls = []
final =[]
website = []
@jit()
def fastmachinecode():
    for i in range(1,295):
        url = 'https://www.appfutura.com/app-developers?p='+str(i)
        reqs = requests.get(url)
        soup = BeautifulSoup(reqs.text, 'html.parser')
        for link in soup.find_all('a'):
            if link.get('href').startswith("https://") == True:
                urls.append(link.get('href'))
                for j in urls:
                    if j not in final:
                        final.append(j)
                        print(i)
                                     
```

```python
titles = []
titles_url = []
@jit()
def titlesfun():
    for i in final:
        page = requests.get(i)
        soup = bs(page.content)
        html= soup.prettify()
        soup = BeautifulSoup(html, "html.parser") 
        spans=soup.find_all('h1', {'class': 'big-title no-mar-top no-mar-bot strong'})
        for span in spans:
            titles.append(span.text.strip())
            titles_url.append(i)
    return titles , titles_url
```

```python
# Extract the websites
website = []
@jit()
def mywebsite():
    for i in range(1,296):
        url = 'https://www.appfutura.com/app-developers/.html?p='+str(i)
        page = requests.get(url)
        soup = bs(page.content)
        html= soup.prettify()
        soup = BeautifulSoup(html, "html.parser") 
        if soup.find_all('span', {'class': 'js-link btn btn-primary-strong btn100 visit-website-tracking'}) == []:
            website.append('')
        else:
            spans = soup.find_all('span', {'class': 'js-link btn btn-primary-strong btn100 visit-website-tracking'})
            for span in spans:
                website.append(span.get('data-url'))

        if soup.find_all('a', {'class': 'btn btn-primary-strong btn100'}) == []:
            website.append('')
        else:
            spans = soup.find_all('a', {'class': 'btn btn-primary-strong btn100'})
            for span in spans:
                website.append('None')
```

```python
# parse the url and extracting the elements
Ratings = []
location = []
founded = []
employee = []
budget = []
offices = []
description = []
company_logo = []
contact_number = []
slogan = []

@jit()
def myelements():
    for i in titles_url:
        page = requests.get(i)
        soup = bs(page.content)
        html= soup.prettify()
        soup = BeautifulSoup(html, "html.parser") 


        if soup.find_all('span', {'itemprop': 'ratingValue'}) == []:
            Ratings.append('-')
        else:
            spans = soup.find_all('span', {'itemprop': 'ratingValue'})
            for span in spans:
                Ratings.append(span.text.strip())


        if soup.find_all('span', {'itemprop': 'address'}) == []:
            location.append('None')
        else:
            spans = soup.find_all('span', {'itemprop': 'address'})
            for span in spans:
                location.append(span.text.strip())  


        if soup.find_all('span', {'itemprop': 'foundingDate'}) == []:
            founded.append('-')
        else:
            spans = soup.find_all('span', {'itemprop': 'foundingDate'})
            for span in spans:
                founded.append(span.text.strip())  


        if soup.find_all('span', {'itemprop': 'value'}) == []:
            employee.append('Nan')
        else:
            spans = soup.find_all('span', {'itemprop': 'value'})
            for span in spans:
                employee.append(span.text.strip())


        if soup.find_all('i', {'class': 'icon-budget'}) == []:
            budget.append('')
        else:
            spans = soup.find_all('i', {'class': 'icon-budget'})
            for span in spans:
                budget.append(span.text.strip())


        if soup.find_all('p', {'class': 'developer-description'}) == []:
            description.append('None')
        else:
            spans = soup.find_all('p', {'class': 'developer-description'})
            for span in spans:
                description.append(span.text.strip())   


        if soup.find_all('div', {'id': 'offices'}) == []:
            offices.append('None')
        else:
            spans = soup.find_all('div', {'id': 'offices'})
            for span in spans:
                offices.append(span.text.replace('Show phone number','').strip())


        my_logo = []
        for link in soup.find_all('img',{'itemprop':'image'}):
            my_logo.append(link.get('src'))                
            logo = []
            for i in my_logo:
                if i.startswith("/uploads"):
                    logo.append(i)                
                    for i in logo:
                        company_logo.append(f'https://www.appfutura.com{i}')


        if soup.find_all('a', {'class': 'link visit-website-tracking'}) == []:
            contact_number.append('Nan')
        else:
            spans = soup.find_all('a', {'class': 'link visit-website-tracking'})
            for span in spans:
                contact_number.append(span.get('data-content'))


        if soup.find_all('h2', {'class': 'normal-title no-mar'}) == []:
            slogan.append('None')
        else:
            spans = soup.find_all('h2', {'class': 'normal-title no-mar'})
            for span in spans:
                slogan.append(span.text.strip())
```
